In this article I will go over backpropagation in neural networks and show where matrices come into play. There are a lot of online resources that show how this works in scalar form.

```python
# A code block
def fib(x):
    if x == 1:
        return 1
    elif x == 2:
        return 1
    p1 = p2 = 1
    for _ in range(x-2):
        temp = p1 + p2
        p1 = p2
        p2 = temp
    return p2
```
